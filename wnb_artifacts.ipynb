# Datasets

```python
REGISTRY_NAME = "Dataset"
COLLECTION_NAME = "object-detection-dataset"
dataset_target_path = f"wandb-registry-{REGISTRY_NAME}/{COLLECTION_NAME}"

TEAM_ENTITY = "baas_group5"
PROJECT = "basketball_tracker"

run = wandb.init(
    entity=TEAM_ENTITY,
    project=PROJECT,
    job_type="publish_dataset",
    reinit=True
)

from roboflow import Roboflow
def download_general_dataset():
    """Download dataset from Roboflow and return path to data.yaml."""
    rf = Roboflow(api_key=os.getenv("ROBOFLOW_API_KEY"))
    project = rf.workspace("workspace-5ujvu").project("basketball-players-fy4c2-vfsuv")
    version = project.version(17)
    dataset = version.download("yolov11")
    return dataset.location

artifact = wandb.Artifact(
    name="general-object-detection",
    type="dataset",
    description="Processed dataset and labels."
)

dataset_path = download_general_dataset()
artifact.add_dir(local_path=dataset_path, name="general_features")

run.log_artifact(artifact)

run.finish()
```
